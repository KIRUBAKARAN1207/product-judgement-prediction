# building the models

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
br,rr=BaggingClassifier(),RandomForestClassifier()
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.3,random_state=0)
```

```python
clas={}
for i in (lr,br,rr,knn):
    clas['models_%s'%i]=i.fit(xtrain,ytrain)
    print('score value for train for %s is %f'%(i,i.score(xtrain,ytrain)))
    print('score value for test for %s is %f'%(i,i.score(xtest,ytest)))
    print('\n')
```

## from above result we choose random forest because it perform better in both test and train so its chosen for further tunning

# TUNNING

```python
params ={'n_estimators':[250,400,500],'criterion':['gini','entropy'],'max_depth':[4,5,None],'min_samples_split':[2,3,4,5],'min_samples_leaf':[2,3,4],'max_features':['auto','sqrt','log2'],'max_samples':[0.7,0.8]}
```

```python
estimator=RandomForestClassifier(random_state=0,n_jobs=-1)
rfcv=GridSearchCV(estimator=estimator,param_grid=params,cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
rfcv.fit(xtrain,ytrain)
```

```python
# getting best params
rfcv.best_params_
```

```python
rcc2=RandomForestClassifier(n_estimators=250,criterion='entropy',max_depth=5,max_features='auto',max_samples=0.8,min_samples_leaf=4,min_samples_split=2)
```

```python
rcc2.fit(xtrain,ytrain)
```

```python
rcc2.score(xtrain,ytrain)
```

```python
rcc2.score(xtest,ytest)
```
